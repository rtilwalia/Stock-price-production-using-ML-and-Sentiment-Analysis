```python
import numpy as np
import pandas as pd
import numpy.random as rd
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn import model_selection
```

```python
data = pd.read_csv(r"C:\Users\tilwa\majorproject\data\TSLA.csv")
#df = data.drop(["Unnamed: 0"], axis = 1, inplace = False)

Row_list =[] 
#Row_list1 = []

# Iterate over each row 
for index, rows in data.iterrows(): 
    # Create list for the current row 
    my_list =[rows['Open'], rows['High'], rows['Low'], rows['Close'], rows['Volume']] 
    #my_list1 = [rows['House overall [kW]']]
    # append the list to the final list 
    Row_list.append(my_list)
    #Row_list1.append(my_list1)
    #result = np.array(Row_list1).flatten().tolist()
    #result1 = np.round(result)
    
print(Row_list[:10])
print(len(Row_list))
#print(result1[:100])
#print(len(result1))
```

```python
trend = []
for row in Row_list:
    if row[3] - row[0] > 0:
        trend.append(1)
    elif row[3] - row[0] < 0:
        trend.append(-1)
    elif row[3] - row[0] == 0:
        trend.append(0)
            
print(trend)
print(len(trend))
          
trendString = []
for row in Row_list:
    if row[3] - row[0] > 0:
        trendString.append('Positive')
    elif row[3] - row[0] < 0:
        trendString.append('Negative')
    elif row[3] - row[0] == 0:
        trendString.append('Neutral')
        
print(trendString)
```

```python
data["Trend"] = trend
data["TrendString"] = trendString
```
